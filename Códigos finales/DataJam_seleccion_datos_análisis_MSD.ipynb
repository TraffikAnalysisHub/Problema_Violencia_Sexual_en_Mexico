```python
#hago prueba de duumies con bp1_5, falta hacer las demas
TMod_cols_base=['ID_VIV','ID_HOG',	'ID_PER',	'ID_DEL','BPCOD']
TMod_cols=['ID_VIV','ID_HOG',	'ID_PER',	'ID_DEL','BPCOD',       #datos generales de la persona y el delito
           'BP1_5','BP1_1','BP1_4','BP1_8' , 'BP1_9','BP7_1','BP1_10_1' , 'BP1_10_2']           #variables de interes 
```

```python
#Filtro por tipo de delito 13 y 14
TMod_Filter=TMod[(TMod['BPCOD']==13) | (TMod['BPCOD']==14)][TMod_cols]
```

```python
TMod_Filter.groupby(['ID_PER']).agg(['count']) #hay 766 personas únicas
```

#### **CREANDO DUMMIES (PARA CADA CATEGORIA)**

```python
base=TMod_Filter[TMod_cols_base]
base

dum0=pd.get_dummies(TMod_Filter['BPCOD'],dummy_na=True,prefix='Tpo_delito')
dum1=pd.get_dummies(TMod_Filter['BP1_1'],dummy_na=True,prefix='BP1_1')
dum2=pd.get_dummies(TMod_Filter['BP1_4'],dummy_na=True,prefix='BP1_4')
dum3=pd.get_dummies(TMod_Filter['BP1_5'],dummy_na=True,prefix='BP1_5')
dum4=pd.get_dummies(TMod_Filter['BP1_8'],dummy_na=True,prefix='BP1_8')
dum5=pd.get_dummies(TMod_Filter['BP1_9'],dummy_na=True,prefix='BP1_9')
dum6=pd.get_dummies(TMod_Filter['BP7_1'],dummy_na=True,prefix='BP7_1')

test=pd.concat([base,dum0,dum1,dum2,dum3,dum4,dum5,dum6],1)

TMod_VF=test.groupby(['ID_VIV','ID_HOG','ID_PER']).agg('sum')
TMod_VF_2=TMod_VF.reset_index()
TMod_VF_2.head(5)
```

```python
#'ID_VIV','ID_HOG','ID_PER',
names_cols=['ID_VIV',	'ID_HOG',	'ID_PER','BPCOD_Tipo_delito','Tpo_delito_13.0','Tpo_delito_14.0','Tpo_delito_nan',
            'BP1_delito_Ene','BP1_delito_Feb','BP1_delito_mar','BP1_delito_abr','BP1_delito_may','BP1_delito_jun','BP1_delito_Jul',
            'BP1_delito_Ago', 'BP1_delito_sep','BP1_delito_oct','BP1_delito_nov','BP1_delito_dic','BP1_delito_no_sabe','BP1_delito_nan',
            'BP1_4_mañana_6_12_am','BP1_4_tarde_12_18', 'BP1_4_noche_18_24','BP1_4_madrugada_01_06','BP1_4_No_sabe_no_responde','BP1_4_No_sabe_no_responde_NAN',
            'BP1_5_calle','BP1_5_casa','BP1_5_trabajo','BP1_5_negocio_establecimiento', 'BP1_5_lugar_publico', 'BP1_5_trans_pub','BP1_5_carretera',
            'BP1_5_otro','BP1_5_no_sabe','BP1_5_nan',
            "BP1_8_acompaniada_si",'BP1_8_acompaniada_no','BP1_8_9','BP1_8_nan',
            "BP1_9_1_num_delin_1","BP1_9_1_num_delin_2","BP1_9_1_num_delin_3","BP1_9_1_num_delin_4","BP1_9_1_num_delin_5","BP1_9_1_num_delin_6",
            "BP1_9_1_num_delin_9","BP1_9_1_num_delin_9_nan",
            'BP7_1_tpo_ofensa_int_viola',"BP7_1_tpo_ofen_hostig", "BP7_1_tpo_ofen_tocamiento", "BP7_1_tpo_ofen_exhib",
            "BP7_1_tpo_ofen_otra","BP7_1_tpo_ofen_no_sabe","BP7_1_NA_tpo_ofen_NA", "BP7_1_nan_tpo_ofen_nan"]
            
cols_eliminar=['BPCOD_Tipo_delito','Tpo_delito_nan','BP1_delito_no_sabe','BP1_delito_nan','BP1_4_No_sabe_no_responde','BP1_4_No_sabe_no_responde_NAN','BP1_5_otro','BP1_5_no_sabe','BP1_5_nan','BP1_8_9','BP1_8_nan',
               "BP1_9_1_num_delin_9","BP1_9_1_num_delin_9_nan","BP7_1_tpo_ofen_otra","BP7_1_tpo_ofen_no_sabe","BP7_1_NA_tpo_ofen_NA", "BP7_1_nan_tpo_ofen_nan"]

TMod_VF_2.columns=names_cols
TMod_VF_3=TMod_VF_2.drop(cols_eliminar,axis=1)
TMod_VF_3
```

```python

cols_vic_tot=['ID_VIV', 'ID_HOG', 'ID_PER','SEXO','EDAD',	'AREAM',	'CVE_ENT',	'NOM_ENT',	'CVE_MUN'	,'NOM_MUN']
VicTot_f=pd.merge(TMod_VF_3,VicTot[cols_vic_tot],how='left',left_on=['ID_VIV',	'ID_HOG',	'ID_PER'],right_on=['ID_VIV',	'ID_HOG',	'ID_PER'])
VicTot_f   #tabla con personas únicas con variables finales 
```

```python
VicTot_f['EDAD_SEG'] = [ 'Adulto_mayor' if x>=60 else ('Adulto_joven' if x<=30 else 'adulto')for x in VicTot_f['EDAD']]
VicTot_f
```

```python
#Sexo = 2 : Mujer
#sexo = 1: Hombre

d1 = pd.get_dummies(VicTot_f['SEXO'],dummy_na=False,prefix='sexo')
d2 = pd.get_dummies(VicTot_f['EDAD_SEG'],dummy_na=False,prefix='EDAD_SEG')
tab1 = VicTot_f.drop(['ID_VIV','EDAD_SEG', 'ID_HOG','NOM_MUN', 'ID_PER','AREAM', 'CVE_ENT','CVE_MUN','SEXO', 'EDAD'],axis = 1)
deefe = pd.concat([tab1,d1,d2], axis =1) 
data_f =deefe.groupby(['NOM_ENT']).agg(['sum'],axis=0)
data_f = data_f.reset_index()
data_f.columns = data_f.columns.droplevel(1)
data_f.head(3)
```

```python
#pegar poblacion entrevistada por municipio
nom_muni = data_f['NOM_ENT']
data_mds = pd.merge(data_f,pop_edo,how = 'left',left_on = 'NOM_ENT',right_on="NOM_ENT")  
data_mds_ = data_mds.drop("NOM_ENT",axis = 1)
data_mds_.head()
```

```python
data_mds_f = data_mds_.div(data_mds_.iloc[:,42],axis = 'index')
data_mds_ff = pd.concat([nom_muni,data_mds_f], axis = 1)
data_mds_ff.shape
```

```python
# Creación DataFrame:
df_numeros = pd.DataFrame(data_mds_ff)
# Guarda datos en CSV:
df_numeros.to_csv(r'C:\Users\Documents\DATA_JAM\MEXICO\data_mds_ff_edo.csv', header=True, index=False)
```
